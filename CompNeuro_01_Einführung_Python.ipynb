## Variablen
Allgemein sind Variablen in der Programmierung Platzhalter, in denen irgendwelche Objekte gespeichert werden können. "Objekte" kann dabei alles mögliche sein: Zahlen, Zeichenketten, Funktionen, Klasseninstanzen, Arrays, Listen,... 

Im Gegensatz zu anderen Programmiersprachen braucht man bei Python nicht vorher anzusagen, welchen Datentyp eine Variable enthält! Das Bedeutet, dass wir problemlos zwischen Integer- und Float-Variablen wechseln können, ohne dass wir Fehlermeldungen bekommen. In der Fachsprache sagen wir: Python ist "dynamically typed" und nicht "statically typed". Weitere Programmiersprachen, die dynamically typed sind, wären JavaScript (z.B. auf Webseiten), Ruby und Perl. Beispiele für statically typed Programmiersprachen wären C, C++, Java und Rust.

Beispiele für Datentypen in Python sind:
- Integer: ```int```, ganze Zahlen (von $-2^{31}$ bis $2^{31}-1$)
- Long Integer: ```long``` noch größere ganze Zahlne (von $-2^{63}$ bis $2^{63}-1$)
- Float: ```float```, Nachkommazahlen (von $2.2
250738585072014^{-308}$ bis $1.7976931348623157^{308}$) (das ist schon ganz schön groß, zum Vergleich: Das Universum hat etwa $10^{90}$ Atome)
- Boolean: ```bool```, Wahrheitswerte ```True``` und ```False``` oder die beiden Zahlen ```1``` und ```0```
- Strings: ```str``` (siehe später) 

````python
x = 4.0
y = 5

print(type(x))
print(type(y))

print(x)
print(y)
print(x+y)  # Addition
print(x-y)  # Subtraktion
print(x*y)  # Multiplikation
print(x/y)  # Division
print(x%y)  # Modulo (Rest beim Teilen)
print(y**0.5) # Potenz (hier: Wurzel)

print(int(22/7))
print(22//7) # Integerdivision (Teilen mit Abrunden)
print(22%7)
print(float(7))
````

## Strings
Strings sind Zeichenketten, also Aneinanderreihung von Buchstaben, Zahlen oder Sonderzeichen. 

Möchte man zwei Strings aneinanderhängen, so kann man sie einfach addieren. Multipliziert man einen String mit einem Integer (Ganzzahl, siehe oben), so wird er entsprechend häufig hintereinander gehängt. (Das ergibt auch Sinn: ```"-"+"-"+"-"``` ist das selbe wie ```3*"-"```.)

````python
name = "Olli"
alter = 32

print(name+ " ist " + str(alter) + " Jahre alt.")
print(20*"-")
````

## Listen
Listen werden genutzt, um mehrere Objekte in einer Variablen zu speichern. Das Praktische hierbei ist, dass es (im Gegensatz zu anderen Programmiersprachen) egal ist, welche Datentypen wir in einer Liste speichern. 

Will man mit einem bestimmten Element einer Liste weiterarbeiten, so muss man lediglich den Index des Elements kennen. Wichtig hierbei ist, dass Python **immer** bei 0 anfängt zu zählen!

Genau wie bei Strings kann man zwei Listen sehr leicht aneinanderhängen, wenn man sie einfach addiert. Ebenfalls eine Multiplikation von einer Liste mit einem Integer ist wie bei Strings möglich.

````python
liste = [1,2,3,4,5,6,7,8,9,"Hallo"] # Liste mit Integern und einem String
print(liste)

print(liste[3]) # Greife auf das vierte Element zu (da wir bei 0 anfangen zu zählen)
print(liste[-5]) # Greife auf das fünftletzte Element zu

# Slicen
print(liste[2:5]) # Teilliste von Element 3 bis (ausschließlich) Element 5
print(liste[:5]) # Teilliste von Element 0 bis (ausschließlich) Element 5
print(liste[5:]) # Teilliste von Element 5 bis (einschließlich) letztem Element
print(liste[2:8:3]) # Teilliste von Element 2 bis (ausschließlich) Element 8 in 3er-Schritten
print(liste[::-1]) # (Teil-)Liste von (einschließlich) Anfang bis Ende, allerdings in -1er-Schritten (also rückwärts)

print(liste+liste) # Hänge liste an liste (konkatenieren, concatenate)
print(3*liste) # Hänge die Liste liste drei mal hintereinander

print(len(liste)) # Gib an, wie viele Elemente in der Liste liste sind
print(len(name)) # Gib an, wie viele Zeichen in dem String name sind

listevonlisten = [liste, liste, liste] # Eine Liste, in der mehrere Listen stecken
print(len(listevonlisten))
````

## For-Schleifen
Schleifen sind Kontrollstrukturen, die iterativ eine Gruppe von Anweisungen mit bestimmter Anzahl von Wiederholungen oder Argumenten ausführen können. Einfach ausgedrückt bedeutet das, dass wir ein iterierbares Objekt (z.B. eine Liste oder einen String) nehmen, jedes Element in diesem Objekt (bei Listen die Einträge, bei Strings die einzelnen Zeichen) durchgehen und entsprechend damit etwas machen.

Die Besonderheit bei Python ist, dass wir immer sogenannte For-Each-Schleifen benutzen. Während wir bei anderen Programmiersprachen meistens einfach Indizes hochzählen, speichern wir die einzelnen Einträge unseres iterierbaren Objektes in einer Variable, die wir dann innerhalb der Schleife verwenden.

Möchten wir jedoch mit Zahlenlisten arbeiten, ohne zuvor eine Liste mit tausenden Werten anzulegen, gibt es das ```range```-Objekt. 

````python
for eintrag in liste[:-1]: # "Nimm die Liste ohne den letzten Eintrag"
    eintrag = (eintrag+1)
    print(eintrag)
print(liste)

n_epoch = 100
for epoch in range(n_epoch): # "Laufe von 0 bis (ausschließlich) n_epoch"
    print(epoch/10)

n_epoch = 100
for epoch in range(3,n_epoch,2): # "Laufe von 3 bis (ausschließlich) n_epoch in 2er-Schritten"
    print(epoch/10)

for i, buchstabe in enumerate("Hallo"): # "Nummeriere jedes Objekt durch"
    if i%2==0:
        continue
    print(buchstabe)
````

## If-Abfragen
If-Abfragen sind Kontrollstrukturen, die den Kontrollfluss unseres Codes steuern können. Grundsätzlich bedeutet das, dass wir bestimmten Code nur dann ausführen, wenn eine bestimmte Bedingung erfüllt ist.

````python
liste1 = []
liste2 = []
liste3 = list()
for i in range(20):
    if i%4==0: # "Wenn die Zahl durch 4 teilbar ist..."
        liste1.append(i)
    elif i%4==1: # "Wenn ansonsten die Zahl beim Teilen durch 4 Rest 1 hat..."
        liste2 = liste2+[-i]
    else: # "Ansonsten..."
        liste3 += [i**2]

print(liste1)
print(liste2)
print(liste3)
````

## Dictionaries

Listen lassen sich (sehr umständlich) als Dictionaries umschreiben, indem man als Key einfach einen Index übergibt. Das ist jedoch sehr unüblich.

## Beispielprojekt: Cellular Automaton
In diesem Beispielprojekt sollen jetzt (fast) alle soeben gelernten Prinzipien angewendet werden.

Ein Cellular Automaton (Zellulärer Automat) dient der Modellierung diskreter dynamischer Systeme, bei denen das System zum Zeitpunkt $t$ direkt von Zeitpunkt $t-1$ abhängt. Es handelt sich dabei um eine wichtige Grundlage für künstliches Leben. Eine genauere Beschreibung findet sich bei [Wikipedia](https://de.wikipedia.org/wiki/Zellul%C3%A4rer_Automat). 

Wir möchten gerne Rule 110 umsetzen: Rule 110 ist ein 1-dimensionaler Zell-Automat, bei dem jede Zelle entweder lebt oder tot ist. Ein Update-Schritt einer Zelle ist hierbei nur abhängig von den direkten Nachbarn. Der genaue Ablauf des Experimentes findet sich ebenfalls bei [Wikipedia](https://en.wikipedia.org/wiki/Rule_110).

````python
import matplotlib.pyplot as plt # Importiere Zeugs zum Plotten

# Rule 110
n_felder = 200
lebensraum = n_felder*[0] # Erstelle Lebensraum der Zellen
lebensraum[100] = 1 # Erwecke eine Zelle zum Leben

print(lebensraum)

# Regel 110: 0*1+1*2+1*4+1*8+0*16+1*32+1*64+0*128=110
regeln = {
          "111": 0, # Lebe ich und meine beiden Nachbarn, sterbe ich
          "110": 1, # Leben mein linker Nachbar und ich, bleibe ich am Leben
          "101": 1, # Bin ich tot und meine beiden Nachbarn leben, erwache ich zum Leben
          "100": 0, # Bin ich tot und mein linker Nachbar lebt, bleibe ich tot
          "011": 1, # Lebe ich und mein rechter Nachbar, bleibe ich am Leben
          "010": 1, # Lebe ich und meine beiden Nachbarn sind tot, bleibe ich am Leben
          "001": 1, # Bin ich tot und nur mein rechter Nachbar lebt, erwache ich zum Leben
          "000": 0  # Sind meine Nachbarn und ich tot, bleibe ich tot
        }

verlauf = []
for epoch in range(400):
    lebensraum_neu = n_felder*[0]
    for i in range(len(lebensraum)):
        if i==n_felder-1:
            update = str(lebensraum[i-1])+str(lebensraum[i])+str(lebensraum[0])
        else:
            update = str(lebensraum[i-1])+str(lebensraum[i])+str(lebensraum[i+1])
        # Alternativ:
        # update = str(lebensraum[i-1])+str(lebensraum[i])+str(lebensraum[(i+1)%len(lebensraum)])
        lebensraum_neu[i] = regeln[update]
    lebensraum = lebensraum_neu
    verlauf.append(lebensraum)

# Plotte den Verlauf des Experimentes
plt.figure()
plt.imshow(verlauf)
plt.show()
````
